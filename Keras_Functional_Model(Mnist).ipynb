```python
import keras
from keras.models import Model
from keras.layers import *
from keras.datasets import mnist
from keras import optimizers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
xtraindf = pd.DataFrame(X_train.reshape(60000, 784))
ytraindf = pd.DataFrame(y_train.reshape(60000, 1))
xtestdf = pd.DataFrame(X_test.reshape(10000, 784))
ytestdf = pd.DataFrame(y_test.reshape(10000, 1))
```

```python
NUM_ROWS = 28
NUM_COLS = 28
NUM_CLASSES = 10
BATCH_SIZE = 128
EPOCHS = 10

# Reshaping the data to single dimensions
X_train = X_train.reshape(60000, 784)
X_train = X_train.astype('float32')/255 # normalize the pixels
X_test = X_test.reshape(10000, 784)
X_test = X_test.astype('float32')/255

# One-hot encoding
y_train = keras.utils.to_categorical(y_train, NUM_CLASSES)
y_test = keras.utils.to_categorical(y_test, NUM_CLASSES)
```

```python
print(y_train[0])
print(y_train[3])
```

```python
# Input Parameters
n_input = 784
n_hidden_1 = 300
n_hidden_2 = 100
n_hidden_3 = 100
n_hidden_4 = 200
n_classes = 10
```

```python
# First Model
Inp = Input(shape=(784,))
x = Dense(n_hidden_1, activation='relu', name="Hidden_Layer_1")(Inp)
x = Dense(n_hidden_2, activation='relu', name="Hidden_Layer_2")(x)
x = Dense(n_hidden_3, activation='relu', name="Hidden_Layer_3")(x)
x = Dense(n_hidden_4, activation='relu', name="Hidden_Layer_4")(x)
output = Dense(n_classes, activation='softmax', name="Output_Layer")(x)
```

```python
# Our model has 6 layers - input layer, 4 hidden layers and 1 output layer
model = Model(Inp, output)
model.summary()
```

```python
# Inserting Hyperparameters
learning_rate = 0.1
training_epochs = 20
batch_size = 100
sgd = optimizers.SGD(learning_rate=learning_rate)
```

```python
# We rely on the plain vanilla Stochastic Gradient Descent as our Optimizing methodology
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# We have build the model now we have to fit the data in the model
history1 = model.fit(X_train, y_train, batch_size=batch_size, epochs=training_epochs, verbose=2, validation_data=(X_test, y_test))
```

```python
# Second Model
Inp = Input(shape=(784,))
x = Dense(n_hidden_1, activation='relu', name="Hidden_Layer_1")(Inp)
x = Dense(n_hidden_2, activation='relu', name="Hidden_Layer_2")(x)
x = Dense(n_hidden_3, activation='relu', name="Hidden_Layer_3")(x)
x = Dense(n_hidden_4, activation='relu', name="Hidden_Layer_4")(x)
output = Dense(n_classes, activation='softmax', name="Output_Layer")(x)

# We rely on Adam as our optimizing methodology
adam = optimizers.Adam(learning_rate=learning_rate)
model2 = Model(Inp, output)
model2.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history2 = model2.fit(X_train, y_train, batch_size=batch_size, epochs=training_epochs, verbose=1, validation_data=(X_test, y_test))
```
